```python
import pandas as pd
import numpy as np
from pandas_profiling import ProfileReport
import plotly.express as px
pd.options.plotting.backend = "plotly"
import ipywidgets
import matplotlib.pyplot as plt
```

```python
sales = pd.read_csv('../data/sales_train.csv',parse_dates=[
    'date'
])
```

```python
sales.head()
```

```python
sales.isnull().sum()
```

# Dates

```python
sales.date.value_counts().sort_index().plot.line(title='Sales per day')
```

# Date Block Num

```python
sales.date_block_num.value_counts().sort_index().plot.line(title='Number of Sales per month')
```

# Shop ID

```python
sales.shop_id.value_counts(normalize=True).to_frame().reset_index().sort_values(
    "shop_id"
).plot.bar(
    x="index", y="shop_id", title="Percentage of sales per store ID"
).update_layout(
    xaxis={"type": "category"}
)
```

# Item ID

```python
sales.item_id.value_counts().head(1000).plot.bar(
    title="Sales of Top 1000 most sold items",
    labels={"value": "sales", "index": "item_id"},
).update_layout(xaxis={"type": "category"})
```

```python
sales.item_id.value_counts().tail(1000).plot.bar(
    title="Sales of Top 1000 least sold items",
    labels={"value": "sales", "index": "item_id"},
).update_layout(xaxis={"type": "category"})
```

```python
sales.item_id.value_counts().value_counts().head(100).plot.bar(
    title="How many items were sold how many times?",
    labels={
        'index':'number of times sold',
        'value':'number of items '
    }
)
```

# Item Price

```python
sales.item_price.quantile(np.linspace(0, 1, 200)).plot.line(
    title="Percentage of sales per item price"
)
```

# Item Count per Day

```python
item_count_cumulative_distribution = sales.item_cnt_day.quantile(
    np.linspace(0, 1, 200)
).cumsum()
```

```python
colors = np.where(item_count_cumulative_distribution > 0, "positive", "negative")
```

```python
px.bar(
    item_count_cumulative_distribution, color=colors, title="Percentage of count usage"
)
```
